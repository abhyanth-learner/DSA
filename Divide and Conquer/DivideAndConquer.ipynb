FIBONACCI

```python
def fibonacci(n):
    if n< 1:
        return
    if n==1:
        return 0
    if n==2:
        return 2
    else:
        return fibonacci(n-1)+fibonacci(n-2)
```

NUMBER fACTOR

```python
def numberFactor(n):
    if n in (0,1,2):
        return 1
    if n==3:
        return 2
    else:
        sub1=numberFactor(n-1)
        sub2=numberFactor(n-3)
        sub3=numberFactor(n-4)
        return sub1+sub2+sub3
```

HOUSE ROBBER

```python
def houseRobber(house,currentIndex):
    if currentIndex >=len(house):
        return 0;
    else:
        firstHouse=house[currentIndex]+houseRobber(house,currentIndex+2)
        skipFirstHouse=houseRobber(house,currentIndex+1)
        return max(firstHouse,skipFirstHouse)
```

CONVERT STING FROM ONE TO ANOTHER

```python
def minStepsToConvert(s1,s2,index1,index2):
    if len(s1)==index1:
        return len(s2)-index2
    if len(s2)==index2:
        return len(s1)-index1
    if  s1[index1]==s2[index2]:
        return minStepsToConvert(s1,s2,index1+1,index2+1)
    else:
        deleteOp=1+minStepsToConvert(s1,s2,index1,index2+1)
        insertOp=1+minStepsToConvert(s1,s2,index1+1,index2)
        replaceOp=1+minStepsToConvert(s1,s2,index1+1,index2+1)
        return min(deleteOp,insertOp,replaceOp)
```

 LONGEST COMMON SUBSEQUECE

```python
def lcs(s1,s2,index1,index2):
    if index1==len(s1) or index2==len(s2):
        return 0
    if s1[index1]==s2[index2]:
        return 1+lcs(s1,s2,index1+1,index2+1)
    else:
        op1=lcs(s1,s2,index1,index2+1)
        op2=lcs(s1,s2,index1+1,index2)
        return max(op1,op2)
```

LONGEST PALINDROMIC  SUBSEQUENCE

```python
def lps(word,startIndex,endIndex):
    if startIndex>endIndex:
        return 0
    if startIndex==endIndex:
        return 1
    if word[startIndex]==word[endIndex]:
        return 2+lps(word,startIndex,endIndex-1)
    else:
        op1=lps(word,startIndex,endIndex-1)
        op2=lps(word,startIndex+1,endIndex)
        return max(op1,op2)
```

MIMIMUM COST TO REACH [0][0] OF 2D ARRAY

```python
def findMinCost(array,row,col):
    if row ==-1 or col==-1:
        return 0
    elif row==0 and col==0:
        return array[0][0]
    else:
        op1=findMinCost(array,row,col-1)
        op2=findMinCost(array,row-1,col)
        return array[row][col]+min(op1,op2)
```

NUMBER OF WAYS WE CAN REACH FIRST CELL OF 2D ARRAY

```python
def noOfWays(array,row,col,cost):
    if cost <0:
        return 0
    elif row==0 and col==0:
        if array[0][0]-cost==0:
            return 1
        else:
            return 0
    elif row==0:
        return noOfWays(array,0,col-1,cost-array[row][col])
    elif col==0:
        return noOfWays(array,row-1,0,cost-array[row][col])
    else:
        op1=noOfWays(array,row-1,col,cost-array[row][col])
        op2=noOfWays(array,row,col-1,cost-array[row][col])
        return op1+op2
```
