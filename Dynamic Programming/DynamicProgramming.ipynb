FIBONACCI TOP DOWN APPROCH MEMORIZATION

```python
def memFibo(n,memory):
    if n==1:
        return 0
    if n==2:
        return 1
    if n not in memory:
        memory[n]=memFibo(n-1,memory)+memFibo(n-2,memory)
    return memory[n]
```

BOTTOM UP APPROCH WITH TABULATION

```python
def fibonacctTab(n):
    tb=[0,1]
    for i in range(2,n+1):
        tb.append(tb[i-1]+tb[i-2])
    return tb[n-1]
```

NUMBER FACTOR TOP DOWN APPROCH

```python
def numberFactorTd(n,mem):
    if n in (0,1,2):
        return 1
    elif n==3:
        return 2
    else:
        if n not in mem:
            sub1=numberFactorTd(n-1,mem)
            sub2=numberFactorTd(n-3,mem)
            sub3=numberFactorTd(n-4,mem)
            mem[n]=sub1+sub2+sub3
        return mem[n]
```

NUMBER FACTOR BOTTOM UP APPROCH

```python
def numberFactorBu(n):
    res=[1,1,1,2]
    for i in range(4,n+1):
        res.append(res[i-1]+res[i-3]+res[i-4])
    return res[n+1]
```

HOUSE ROBBER TOP DOWN

```python
def houseRobberTd(house,currentHouse,mem):
    if currentHouse>=len(house):
        return 0
    else:
        if currentHouse not in mem:
            firstHouse=currentHouse+houseRobberTd(house,currentHouse+2,mem)
            skipFirstHouse=houseRobberTd(house,currentHouse+1,mem)
            mem[currentHouse]=max(firstHouse,skipFirstHouse)
        return mem[currentHouse]
```
