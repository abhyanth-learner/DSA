COIN CHANGE PROBLEM

```python
def coinChange(coins,amount):
    N=amount
    coins.sort()
    index=len(coins)-1
    while True:
        coinValue=coins[index]
        if N>=coinValue:
            print(coinValue)
            N=N-coinValue
        if(N<coinValue):
            index-=1
        if N==0:
            break
```

KNAPSACK PROBLM

```python
class item:
    def __init__(self,weight,value):
        self.weight=weight
        self.value=value
        self.ratio=value/weight
def knapsackAlgo(items,capacity):
    items.sort(key=lambda x:x.ratio,reverse=True)
    usedCapacity=0
    totalValue=0
    for i in items:
        if usedCapacity+i.weight<=capacity:
            usedCapacity+=i.weight
            totalValue+=i.value
        else:
            unusedWeight=capacity-usedCapacity
            totalValue+=i.ratio*unusedWeight
            usedCapacity+=unusedWeight
    print(totalValue)
```
