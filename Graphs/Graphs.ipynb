```python
import heapq
class Edge:
    def __init__(self,weight,start_vertex,target_vertex):
        self.weight=weight
        self.start_vertex=start_vertex
        self.target_vertex=target_vertex
class Node:
    def __init(self,name):
        self.name=name
        self.min_distance=float('inf')
        self.neighbours=[]
        self.predecessor=None
        self.visited=False
    def __lt__(self,other_node):
        return self.min_distance<other_node.min_distance 
    def add_edge(self,weight,destination_vertex):
        edge=Edge(weight,self,destination_vertex)
        self.neighbours.append(edge)

class Dijkstra():
    def __init__(self):
        self.heap=[]
    def calculate(self,start_vertex):
        start_vertex.min_distance=0
        heapq.heappush(self.heap,start_vertex)
        while self.heap:
            actual_vertex=heapq.heappop(self.heap)
            if(actual_vertex.visited):
                continue
            for edge in actual_vertex.neighbours:
                start=edge.start_vertex
                end=edge.target_vertex
                new_distance=edge.weight+start.min_distance
                if new_distance <edge.min_distance:
                    end.min_distance=new_distance
                    end.predecessor=start
                    heapq.heappush(self.heap,end)
            actual_vertex.visited=True
    def get_Shortest_path(self,vertex):
        print(f"The shortest Path to the verte is:{vertex.min_distance}")
        actual_vertex=vertex
        while actual_vertex is not None:
            print(actual_vertex.name,end=" ")
            actual_vertex=actual_vertex.predecessor
```

FLOYD WARSHALL ALGORITHM

```python
INF=9999
def printSolution(nV,dist):
    for i in range (nV):
        for j in range(nV):
            if(dist[i][j]==INF):
                print("INF",end=" ")
            else:
                print(dist[i][j],end=" ")
        print(" ")

def floydWarshall(nV,G):
    distance=G
    for k in range (nV):
        for i in range (nV):
            for j in range(nV):
                distance[i][j]=min(distance[i][j],distance[i][k]+distance[k][j])
    print(printSolution(nV,distance))
```

DISJOINT SET

```python
class disjointSet:
    def __init__(self,vertices):
        self.vertices=vertices
        self.parent={}
        for v in vertices:
            self.parent[v]=v
        self.rank=dict.fromkeys(vertices,0)
    
    def find(self,item):
        if  self.parent[item]==item:
            return item
        else:
            return self.find(self.parent[item])
        
    def union(self,x,y):
        xroot=self.find(x)
        yroot=self.find(y)
        if self.rank[xroot] < self.rank[yroot]:
            self.parent[xroot] = yroot
        elif self.rank[xroot] > self.rank[yroot]:
            self.parent[yroot] = xroot
        else:
            self.parent[yroot] = xroot
            self.rank[xroot] += 1
```

KRUSKAL'S ALGORITHM

```python
class kGraph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = []
        self.nodes = []
        self.MST = []

    def addEdge(self, s, d, w):
        self.graph.append([s, d, w])
    
    def addNode(self, value):
        self.nodes.append(value)
    
    def printSolution(self,s,d,w):
        for s, d, w in self.MST:
            print("%s - %s: %s" % (s, d, w))

    def kruskalAlgo(self):
        start,end=0,0
        ds=disjointSet(self.nodes)
        self.graph=sorted(self.graph,key=lambda item:item[2])
        while end <self.V-1:
            s,d,w=self.graph[start]
            start+=1
            x=ds.find(s)
            y=ds.find(d)
            if(x!=y):
                end+=1
                self.MST.append([s,d,w])
                ds.union(x,y)
        self.printSolution(s,d,w)
```
