```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
    def __str__(self):
        return str(self.value)
```

```python
class LinkedList:
    def __init__ (self):
        self.head=None
        self.tail=None
        self.length=0
    def __str__(self):
        current=self.head
        res=""
        while current is not None:
            res+=str(current.value)
            if current.next:
                res+=" -> "
            current=current.next
        return res
    def append(self,value):
        new_node=Node(value)
        if self.head==None:
            self.head=new_node
            self.tail=new_node
        else:
            self.tail.next=new_node
            self.tail=new_node
        self.length+=1
    def prepend(self,value):
        new_node=Node(value)
        if self.head==None:
            self.head=new_node
            self.tail=new_node
        else:
            new_node.next=self.head
            self.head=new_node
        self.length+=1
    def insert(self,index,value):
        if(self.head is None):
            self.head=new_node
            self.tail=new_node
        elif(index==(self.length-1)):
            self.append(value)  
        elif(index==0):
            self.append(value)
        else:
            new_node=Node(value)  
            current=self.head    
            for i in range(index-1):
                current=current.next
            new_node.next=current.next
            current.next=new_node
        self.length+=1
    def search(self,value):
        current=self.head
        for i in range(self.length):
            if(current.value==value):
                return i
            current=current.next
        return -1
    def get (self,index):
        if(index>=self.length):
            return None
        else:
            current=self.head
            for i in range (index):
                current=current.next
            return current
    def set(self,value,index):
        current=self.get(index)
        if(index>=self.length):
            return None
        elif(current):
            current.value=value
    def pop_first(self):
        if(self.length==0):
            return None
        popped_node=self.head
        if(self.length==1):
            self.head=None
            self.tail=None
        else:
            self.head=self.head.next
            popped_node.next=None
            
        self.length-=1
        return popped_node
    def pop(self):
        if self.length == 0:
            return None
        popped_node = self.tail
        if self.length == 1:
            self.head = self.tail = None
        else:
            temp = self.head
            while temp.next is not self.tail:
                temp = temp.next
            temp.next = None
            self.tail = temp
        self.length -= 1
        return popped_node 
    def remove(self,index):
            prev=self.get(index-1)
            popped_node=prev.next
            prev.next=prev.next.next
            return popped_node
    def delete_all(self):
        self.head=None
        self.tail=None
        self.length=0
    def getTheMiddel(self):
        fast=self.head
        slow=self.head
        while fast is not None and fast.next is not None:
            fast=fast.next.next
            slow=slow.next
        return slow.value 
    def removeDuplicates(self):
        current=self.head
        vals=set()
        index=1
        vals.add(current.value)
        while current.next is not None:
            if(current.next.value in vals):
                self.remove(index)
            else:
                current=current.next
                index+=1
    def reverse(self):
        curr=self.head 
        prev=None
        while curr is not None:
            next=curr.next
            curr.next=prev
            prev=curr
            curr=next
        self.head,self.tail=self.tail,self.head
        return prev
    def isPalindrome(self, head):
        # TODO
        l1=list()
        cur=head
        while cur is not None:
            l1.append(cur.val)
            cur=cur.next
        i=0
        j=-1
        flag=True
        while i<len(l1):
            if(l1[i]!=l1[j]):
                flag=False  
                break
            i+=1
            j-=1
        return flag
```

DOUBLEY LINKED LIST

```python
class NodeD:
    def __init__(self,value):
        self.value=value
        self.prev=None
        self.next=None
    def __str__(self):
        return str(self.value)
```

```python
class DoublyLinkedList:
    def __init__(self):
        self.head=None 
        self.tail=None
        self.length=0
    def __str__(self):
        current_node=self.head
        res=''
        while current_node:
            res+=str(current_node.value)
            if current_node.next:
                res+="<->"
            current_node=current_node.next
        return res
    def prepend(self,value):
        new_node=NodeD(value)
        if(self.length==0):
            self.head=new_node
            self.tail=new_node
        else:
            new_node.next=self.head
            self.head.prev=new_node
            self.head=new_node
        self.length+=1
    def append(self,value):
        new_node=NodeD(value)
        if(self.length==0):
            self.head=new_node
            self.tail=new_node
        else:
            self.tail.next=new_node
            new_node.prev=self.tail
            self.tail=new_node
        self.length+=1
    def get(self,index):
        if(index<0 or index>=self.length):
            return Exception("Index out of bounds")
        elif(index<(self.length//2)):
            current_node=self.head
            for _ in range (index):
                current_node=current_node.next            
        else:
            current_node=self.tail
            for _ in range(self.length-1,index,-1):
                current_node=current_node.prev
        return current_node
    def set(self,value,index):
        current_node=self.get(index)
        if(current_node):
            current_node.value=value
            return True
        return Exception("Index out of bounds")
    def insert(self,value,index):
        if(index<0 or index>self.length):
            return False
        elif(index==0):
            self.prepend(value)
        elif(index==self.length):
            self.append(value)
        else:
            prev_node=self.get(index-1)
            new_node=NodeD(value)
            new_node.next=prev_node.next
            new_node.prev=prev_node
            prev_node.next.prev=new_node
            prev_node.next=new_node
            self.length+=1
    def search(self,value):
        current=self.head
        while current:
            if(current.value==value):
                return True
            current=current.next
        return False
           
           
    def pop_first(self):
        if(self.length==0):
            return None
        popped_node=self.head
        if(self.length==1):
            self.head=None
            self.tail=None
        else:
            popped_node.next.prev=None
            self.head=popped_node.next
            popped_node.next=None
        self.length-=1
        return popped_node
    def pop(self):
        if(self.length==0):
            return None
        popped_node=self.tail
        if(self.length==1):
            self.head=None
            self.tail=None
        else:
            popped_node.prev.next=None
            self.tail=popped_node.prev
            popped_node.prev=None
        self.length-=1
        return popped_node
       
    def remove(self,index):
        if(self.length==0):
            return None
        elif(index==0):
            self.pop_first()
        elif(index==self.lenth-1):
            self.pop()
        else:
            popped_node=self.get(index)
            popped_node.next.prev=popped_node.prev
            popped_node.prev.next=popped_node.next
            popped_node.next=None
            popped_node.prev=None
        self.length-=1
        return popped_node
    def deleteDLL(self):
        self.head=None
        self.tail=None
```

```python
class CircularDoubleyLiskedList:
    def __init__(self):
        self.head=None
        self.tail=None
        self.length=0
    def __str__(self):
        current=self.head
        res=''
        while(current):
            res+=str(current.value)
            current=current.next
            if(current==self.head):
                break
            res+='<->'
        return res
    def prepend(self,value):
        new_node=NodeD(value)
        if(self.length==0):
            self.head=new_node
            self.tail=new_node
            self.head.prev=self.tail
            self.tail.next=self.head
        else:
            new_node.next=self.head
            new_node.prev=self.tail
            self.head.prev=new_node
            self.head=self.head.prev
            self.tail.next=self.head
        self.length+=1
    def append(self,value):
        new_node=NodeD(value)
        if(self.length==0):
            self.head=new_node
            self.tail=new_node
            self.head.prev=self.tail
            self.tail.next=self.head
        else:
            new_node.prev=self.tail
            new_node.next=self.head
            self.tail.next=new_node
            self.tail=self.tail.next
            self.head.prev=self.tail
        self.length+=1
    def get(self,index):
        if(index>=self.length or index<0):
            return None
        if(self.length==0):
            return None
        elif(index<self.length//2):
            current=self.head
            for _ in range(index):
                current=current.next
        else:
            current=self.tail
            for _ in range(self.length-1,index,-1):
                current=current.prev
        return current
    def set(self,index,value):
        current=self.get(index)
        if(current):
            current.value=value
            return True
        return False
    def insert(self,index,value):
        if(index<0 or index>self.length):
            return Exception("Index out of bounds")
        new_node=NodeD(value)
        if(index==0):
            self.prepend(value)
        elif(index==self.length):
            self.append(value)
        else:
            prev_node=self.get(index-1)
            new_node.next=prev_node.next
            new_node.prev=prev_node
            prev_node.next=new_node
            new_node.next.prev=new_node
            self.length+=1
    def search(self,value):
        if(self.length==0):
            return Exception("Empty")
        else:
            current=self.head
            while(current):
                if(current.value==value):
                    return True
                current=current.next
                if(current==self.head):
                    return False
    def pop_first(self):
        if(self.length==0):
            return None
        else:
            popped_node=self.head
            self.head=popped_node.next
            popped_node.next=None
            popped_node.prev=None
            self.tail.next=self.head
            self.head.prev=self.tail
        self.length-=1
        return popped_node
    def pop(self):
        if(self.length==0):
            return None
        else:
            popped_node=self.tail
            self.tail=popped_node.prev
            popped_node.next=None
            popped_node.prev=None
            self.tail.next=self.head
            self.head.prev=self.tail
        self.length-=1
        return popped_node
    def remove(self,index):
        if(index<0 or index>=self.length):
            return Exception("Index ou of range")
        elif(index==0):
            return self.pop_first()
        elif(index==self.length-1):
            return self.pop()
        else:
            prev_node=self.get(index-1)
            popped_node=prev_node.next
            prev_node.next=popped_node.next
            popped_node.next.prev=prev_node
            popped_node.next=None
            popped_node.prev=None
            self.length-=1
            return popped_node
    def delete_all(self):
        self.head=None
        self.tail=None
        self.length=0
```
