Searching Algorithms

LINEAR SEARCH

```python
def linearSearch(customList,val):
    for i in range(len(customList)):
        if(customList[i]==val):
            return i
    return -1

```

BINARY SEARCH

```python
import math
def binarySearch(customList,val):
    start=0
    last=len(customList)-1
    middle=math.floor((last-start)/2)

    while not(middle==val) and start<=last:
        if(val<customList[middle]):
            last=middle-1
        else:
            start=middle+1
        middle=math.floor((last-start)/2)
    if(customList[middle]==val):
        return middle
    else:
        return -1
```
